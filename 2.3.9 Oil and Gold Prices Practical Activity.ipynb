# What is the daily average price of gold and oil in British Pound (GBP)?

```python
def av_price(x, y, z):
    av=(x+y+z)/3
    return av

# Insert avg into a new column.
gold_subset500['average_gold'] = av_price(gold_subset500['Open'], gold_subset500['High'], gold_subset500['Low'])

# View the DataFrame.
gold_subset500
```

```python
# Insert avg into a new column.
oil_subset500['average_oil'] = av_price(oil_subset500['Open'], oil_subset500['High'], oil_subset500['Low'])

# View the DataFrame.
oil_subset500
```

# How does the daily average price (GBP) of gold and oil compare?

```python
# Employ lambda function to convert US dollars to British pounds.
# Where 1 US$ = 0.8 GBP.
gold_subset500['avg_gold_GBP'] = gold_subset500['average_gold'].apply(lambda x: x*0.8)

# View the DataFrame.
gold_subset500
```

```python
# Employ lambda function to convert US dollars to British pounds.
# Where 1 US$ = 0.8 GBP.
oil_subset500['avg_oil_GBP'] = oil_subset500['average_oil'].apply(lambda x: x*0.8)

# View the DataFrame.
oil_subset500
```

```python
# Use the merge function and Date as ID.
gold = gold_subset500[['Date', 'avg_gold_GBP']]

gold
```

```python
# Use the merge function and Date as ID.
oil = oil_subset500[['Date', 'avg_oil_GBP']]

oil
```

```python
gold_oil = pd.merge(gold, oil, how = 'left', on = 'Date')

gold_oil
```
